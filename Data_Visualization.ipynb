**5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(42)
x = np.random.rand(100)
y1 = x + np.random.normal(0, 0.1, 100)
y2 = x * 0.5 + np.random.normal(0, 0.1, 100)
y3 = x * 2 + np.random.normal(0, 0.1, 100)
data = pd.DataFrame({
    'Feature_1': x,
    'Feature_2': y1,
    'Feature_3': y2,
    'Feature_4': y3
})
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# BOKEH ASSIGNMENT

**1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x**

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title='Sine Wave',
           x_axis_label='X',
           y_axis_label='sin(X)',
           sizing_mode="stretch_both")
p.line(x, y, line_width=2, line_color='blue', legend_label='sin(x)')
output_notebook()
show(p)
```

**2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.**

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.randint(5, 50, n)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)
data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
}
df = pd.DataFrame(data)
source = ColumnDataSource(df)
p = figure(title='Random Scatter Plot',
           x_axis_label='X',
           y_axis_label='Y',
           sizing_mode="stretch_both")
p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)
output_notebook()
show(p)
```

**3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))
p = figure(x_range=fruits, title='Fruit Counts',
           x_axis_label='Fruits',
           y_axis_label='Counts',
           sizing_mode="stretch_both")
p.vbar(x='fruits', top='counts', source=source, width=0.9, color='blue')
output_notebook()
show(p)
```

**4. Create a Bokeh histogram to visualize the distribution of the given data.data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)**

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
np.random.seed(42)
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title='Histogram of Random Data',
           x_axis_label='Value',
           y_axis_label='Counts',
           sizing_mode="stretch_both")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black')
output_notebook()
show(p)
```

**5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)**

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar, LinearColorMapper, ColumnDataSource
from bokeh.transform import transform
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
heatmap_data = data_heatmap.flatten()
x_coords = np.repeat(x, len(y))
y_coords = np.tile(y, len(x))
source = ColumnDataSource(data=dict(heatmap_data=heatmap_data, x=x_coords, y=y_coords))
p = figure(title='Heatmap of Random Data',
           x_axis_label='X',
           y_axis_label='Y',
           x_range=(0, 1),
           y_range=(0, 1),
           sizing_mode="stretch_both")
mapper = LinearColorMapper(palette="Viridis256", low=heatmap_data.min(), high=heatmap_data.max())
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        line_color=None, fill_color=transform('heatmap_data', mapper))
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')
output_notebook()
show(p)
```
